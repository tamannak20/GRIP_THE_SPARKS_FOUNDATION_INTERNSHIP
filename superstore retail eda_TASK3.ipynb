<p style = 'font-size:20px;color:black'> <b>removing outlier from the data set using IQR technique</b> </p>

```python
def remove_outliers(colname):
    Q1 = np.percentile(df[colname],25,interpolation='midpoint')
    Q3 = np.percentile(df[colname],75,interpolation='midpoint')
    IQR = Q3-Q1
    print(f'first quartile : {Q1} \nthird quartile: {Q3}\nInter quartile range : {IQR}')
    print('-'*50)
    data_filter= (df[colname] >= Q1 - 1.5 * IQR) & (df_statistics[colname] <= Q3 + 1.5 * IQR)
    data_no_outlier = df[data_filter][colname]
    print(data_no_outlier)
    print('-'*50)
    print(f'checking null values if any present in the column {colname} : {data_no_outlier.isna().sum()}')
    
    sns.boxplot(data=data_no_outlier)
    plt.xlabel(colname)
remove_outliers('Sales')
    
    
```

```python
sns.barplot(palette='Set2',data=df,x='Category',y='Sales', hue='Region')
plt.show()
```

```python
subcat_sale = round(df.groupby('Sub-Category')['Sales'].sum(),2)
subcat_sale_reset= subcat_sale.reset_index()
```

```python
subcat_sale_df = subcat_sale_reset.pivot_table(index=None,columns='Sub-Category',values="Sales",aggfunc='sum')
```

```python
subcat_profit = round((df.groupby('Sub-Category')['Profit'].sum()/df['Profit'].sum())*100,2)
subcat_profit_reset= subcat_profit.reset_index()
subcat_profit_df = subcat_profit_reset.pivot_table(index=None,columns='Sub-Category',values="Profit",aggfunc='sum')
subcat_Quantity = round(df.groupby('Sub-Category')['Quantity'].sum(),2)
subcat_Quantity_reset= subcat_Quantity.reset_index()
subcat_Quantity_df = subcat_Quantity_reset.pivot_table(index=None,columns='Sub-Category',values="Quantity",aggfunc='sum')
```

```python
df_subcat = pd.concat([subcat_sale_df,subcat_profit_df,subcat_Quantity_df],axis=0)
df_subcat
```

```python
sns.countplot(data=df,x='Quantity',hue='Segment')
plt.legend(loc='upper right')
plt.show()
```

```python
#obtaining percentage of each catgeory 
office_supplies_sales = df[df['Category'] == 'Office Supplies']['Sales']
Furniture_sales = df[df['Category'] == 'Furniture']['Sales']
Technology_sales = df[df['Category'] == 'Technology']['Sales']
total_sale = df["Sales"].sum()
perc_off_supp = (office_supplies_sales/total_sale)*100
perc_furniture = (Furniture_sales/total_sale)*100
perc_Tech_supp = (Technology_sales/total_sale)*100
off_sup = round(perc_furniture.sum(),2)
furniture_ = round(perc_off_supp.sum(),2)
tech_= round(perc_Tech_supp.sum(),2)
data = [off_sup,furniture_,tech_]
category =['Office_supply','furniture','tech']
explode = (0, 0.05, 0.05)
plt.pie(data,labels=category,autopct='%1.1f%%',explode=explode,colors=sns.color_palette('Set2'),pctdistance=1.2,labeldistance=0.4)
plt.title('sale breakage category wise ')
plt.show()
```

```python
# which category is most profitable and is having highest sell 
cat_prof = round((df.groupby(['Category'])['Profit'].sum()/df['Profit'].sum())*100,2)
cat_sales = round(df.groupby(['Category'])['Sales'].sum(),2)
cat_pro_sale = pd.concat([cat_prof,cat_sales],axis=1)
cat_pro_sale
```
